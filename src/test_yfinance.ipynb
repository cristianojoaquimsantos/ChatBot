```python
def retorna_cotacao(ticker_id, period):
    ticker_obj_function = yf.Ticker(f"{ticker_id}.SA")
    hist = ticker_obj_function.history(period=period)["Close"]
    hist.index = hist.index.strftime("%Y-%m-%d")
    hist = round(hist, 2)

    if len(hist) > 30:
        slice_size = int(len(hist) / 30)
        hist = hist[::-slice_size][::-1]

    return hist.to_json()
```

```python
hist_data = retorna_cotacao("BBDC4", "5d")
hist_data
```
